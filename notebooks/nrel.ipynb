```python
import os
import pandas as pd
```

# SKILLS

```python
path = '/Users/juanignaciosolerno/Downloads/julius/datasets/nrel_august_data_ready_for_phrasent.csv'
if os.path.isabs(path):
    df = pd.read_csv(path,dtype='str')
    print("DataFrame cargado correctamente.")
else:
    print("La ruta proporcionada no es absoluta.")
```

```python
df[df['skills']=='']
```

```python
df[df['skills'].isna()]
```

```python
grouped_df = df.groupby('parsed_jt').agg(
    unique_id_count=('id', 'nunique'),
    associated_reference_jt_list=('reference_jt', lambda x: list(x.unique()))
).reset_index()
```

```python
grouped_df['associated_reference_jt_list'] = grouped_df['associated_reference_jt_list'].apply(lambda x: ','.join(x))
```

```python
grouped_df = grouped_df.astype(str)
grouped_df.sample(30)
```

```python
path='../data/processed/nrel/id_count_by_parsed_jt_skills.csv'
grouped_df.to_csv(path,index=False)
```

# CERTIFICATIONS

```python
path = '/Users/juanignaciosolerno/Downloads/julius/datasets/nrel_august_all_data_for_certifications_models_granular.csv'
df2 = pd.read_csv(path,dtype='str')
df2.sample()
```

```python
df2[(df2['certification']=='') | (df2['certification'].isna())]
```

```python
df_final = pd.concat(
    [df[['id','parsed_jt','reference_jt']].drop_duplicates(),
     df2[['id','parsed_jt','reference_jt']].drop_duplicates(),
     df3[['id','parsed_jt','reference_jt']].drop_duplicates()]
).drop_duplicates().reset_index(drop=True)
```

```python
df_final[df_final['reference_jt'].str.len() == 1]
```

```python
grouped_df_final= df_final.groupby('parsed_jt').agg(
    unique_id_count=('id', 'nunique'),
    associated_reference_jt_list=('reference_jt', lambda x: list(x.unique()))
).reset_index()
grouped_df_final['associated_reference_jt_list'] = grouped_df_final['associated_reference_jt_list'].apply(lambda x: [item.lower() for item in x])
grouped_df_final.sample(20)
```
